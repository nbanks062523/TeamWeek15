```python
import pandas as pd 

# Load the cleaned data from each csv file into a pandas dataframe
obesity_df = pd.read_csv('./dsa-airflow/data/obesity_rating_cleaned.csv', header=0)
subway_df = pd.read_csv('./dsa-airflow/data/subway_stores_cleaned.csv', header=0)
chipotle_df = pd.read_csv('./dsa-airflow/data/chipotle_stores_cleaned.csv', header=0)
```

```python
subway_series = subway_df['state'].value_counts().reset_index()
subway_series.columns = ['state_abbrev', 'subway_count_per_state']
#display(subway_series)
obesity_subway_df = obesity_df.join(subway_series.set_index('state_abbrev'), on='state_abbrev')
display(obesity_subway_df)

```

```python
chipotle_series = chipotle_df['state'].value_counts().reset_index()
chipotle_series.columns = ['state', 'chipotle_count_per_state']
#display(chipotle_series)
obesity_chipotle_subway_df = obesity_subway_df.join(chipotle_series.set_index('state'), on='state')
#display(obesity_chipotle_subway_df.isna().sum())
obesity_chipotle_subway_df.fillna(0, inplace=True)
display(obesity_chipotle_subway_df)
```

```python
import numpy as np

obesity_chipotle_subway_df['sq_mi_per_subway'] = obesity_chipotle_subway_df['area_sq_mi'] / obesity_chipotle_subway_df['subway_count_per_state']
obesity_chipotle_subway_df['sq_mi_per_chipotle'] = obesity_chipotle_subway_df['area_sq_mi'] / obesity_chipotle_subway_df['chipotle_count_per_state']
obesity_chipotle_subway_df = obesity_chipotle_subway_df.round({'sq_mi_per_subway': 0, 'sq_mi_per_chipotle': 0})
obesity_chipotle_subway_df = obesity_chipotle_subway_df.replace([np.inf, -np.inf], np.nan)
display(obesity_chipotle_subway_df)
```

```python
display(obesity_chipotle_subway_df.dtypes)
obesity_chipotle_subway_df.to_csv('./dsa-airflow/data/obesity_rating_final.csv', index=False)
```
